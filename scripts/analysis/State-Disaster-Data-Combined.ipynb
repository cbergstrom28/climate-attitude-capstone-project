```r
# Get a list of all files ending with ".csv" in the current directory
files_list <- list.files(pattern = "*.csv",full.names = TRUE)

print(files_list)
```

```r
# Read all files and combine them into a single data frame
all_data <- set_names(files_list) %>%
map_dfr(read_csv, skip = 2, .id = "source_file") %>%
mutate(
state = str_extract(basename(source_file), "(?<=events-)[A-Z]{2}")
)
# View the combined data
head(all_data)
```

```r
#filter for just the hurricane data and the columns being used

hurricane_data <- all_data %>%
    filter(Disaster =="Tropical Cyclone") %>%
    select(state, Name, "Begin Date", "CPI-Adjusted Cost", Deaths) %>%
filter(`CPI-Adjusted Cost` > 50000)


hurricane_data
```
